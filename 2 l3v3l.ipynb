```julia
using LinearAlgebra 
using Kronecker
using MKL
using Plots, PyPlot
using Random, Distributions
using RandomMatrices
```

```julia
#The spin 1 Algebra



#The X,Y,Z gates:


Z = [1 0 0; 0 0 0; 0 0 -1 ];
X = [0 1 0 ; 1 0 1; 0 1 0 ]/(sqrt(2));
Y = [0 -1im 0 ; im 0 -im; 0 im 0]/(sqrt(2));
S=  [0 1 0 ; 0 0 1; 1 0 0 ];

#The finite X,Y,Z Rotations:


function RX(r)
    exp(-im*r/2*X)
end


function RY(r)
    exp(-im*r/2*Y)
end


function RZ(r)
    exp(-im*r/2*Z)
end
```

```julia
FU=fill(fill(0.1+im, 9,9), 5);

        


fone=copy(kronecker(Z,Z))
    for j in 1:length(FU)
    FU[j]=exp(-im*J[j]*fone);
    end


    
    UOdd=copy(kronecker(FU[1],FU[2],FU[3]));
    UEven=copy(kronecker(I(3),FU[4],FU[5],I(3)));


    A=XRow*UEven*UOdd
    
   
 
    
    
EigA=eigvals(A);

N=fill(0.1,length(EigA))
N=angle.(EigA);

N= sort(N,rev=true);


EigA1=copy(N);
ls = deleteat!(EigA1,1);
la = deleteat!(N,length(N));
m=copy(la-ls)/mean(la-ls)

histogram(m)

```
